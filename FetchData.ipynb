```python
def ticker_to_api_ticker(ticker):
    split_tick = ticker.split(":")
    if len(split_tick) == 2:
        return split_tick[1] + "." + split_tick[0]
    elif len(split_tick) == 1:
        return ticker
    else:
        print("PROBLEM")

print(ticker_to_api_ticker("TSX:SSL"))
print(ticker_to_api_ticker("GOLD"))
```

```python
names_and_tickers_list = [(val[0], ticker_to_api_ticker(val[1])) for val in names_and_tickers_list]
```

```python
with open("api_key.txt") as api_key_f:
    api_key = api_key_f.readline()
api_url = "https://financialmodelingprep.com/api/v3"
```

```python
ratio_ttm_api_url = api_url + '/ratios-ttm'
```

```python
ratio_ttm_api_url
```

### Remove tickers that occur in two markets

```python
whitelist = set('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')
names_and_tickers_list = []
for ind, name in enumerate(names_list):
    name_chars = ''.join(filter(whitelist.__contains__, name))
    last_name_chars = ''.join(filter(whitelist.__contains__, names_list[ind-1]))
    if ind > 0 and name_chars.upper() == last_name_chars.upper():
        continue
    names_and_tickers_list.append((name, ticker_to_api_ticker(tickers_list[ind])))
names_and_tickers_list
print(len(names_and_tickers_list))
```
