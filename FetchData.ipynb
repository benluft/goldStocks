```python
def ticker_to_api_ticker(ticker):
    split_tick_colon = ticker.split(":")
    split_tick_dot = ticker.split('.')
    if len(split_tick_colon) == 2:
        if split_tick_colon[0] in ["TSX"]:
            return split_tick_colon[1] + "." + "TO"
        elif split_tick_colon[0] in ["V", "CSE"]:
            return None
        else:
            print(f"BEN YOU DIDN'T ACCOUNT FOR THIS COLON ENDING {split_tick_colon[0]}")
    elif len(split_tick_dot) == 2:
        if split_tick_dot[1] in ['V']:
            return None
        elif split_tick_dot[1] not in ['TO']:
            print(f"BEN YOU DIDN'T ACCOUNT FOR THIS DOT ENDING {split_tick_dot[1]}")
    elif len(split_tick_colon) == 1:
        return ticker
    else:
        print("PROBLEM")
print(ticker_to_api_ticker("CSE:RLG"))
print(ticker_to_api_ticker("V:VRR"))
print(ticker_to_api_ticker("TSX:SSL"))
print(ticker_to_api_ticker("GOLD"))
```

### Remove tickers that occur in two markets

```python
whitelist = set('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')
names_and_tickers_list = []
for ind, name in enumerate(names_list):
    name_chars = ''.join(filter(whitelist.__contains__, name))
    last_name_chars = ''.join(filter(whitelist.__contains__, names_list[ind-1]))
    if ind > 0 and name_chars.upper() == last_name_chars.upper():
        continue
    ticker_formatted = ticker_to_api_ticker(tickers_list[ind])
    if ticker_formatted:
        names_and_tickers_list.append((name, ticker_formatted))
names_and_tickers_list.append(("Torex Gold", "TXG.TO"))
print(names_and_tickers_list)
print(len(names_and_tickers_list))
```

### Request a single ticker

```python
req = requests.get(ratio_ttm_api_url + f'/GOLD?apikey={api_key}')
```

```python
print(req.json())
print(req.status_code)
```

```python
redo_ttm_jsons = True
if redo_ttm_jsons:
    ratio_ttm_jsons = {}
    for name_ticker in names_and_tickers_list:
        ticker = name_ticker[1]
        call_url = ratio_ttm_api_url + f'/{ticker}?apikey={api_key}'
        print(ticker)
        print(call_url)
        print(f"Response code {req.status_code}")
        req = requests.get(call_url)
        if req:
            ratio_ttm_jsons[f"{ticker}"] = req.json()
        else:
            print(f"Issue with {ticker}")

    with open(os.path.join(os.getcwd(), 'jsons', 'goldStockRatiosTTM.json'), 'w') as f:
        json.dump(ratio_ttm_jsons, f, indent=4)
    
```
